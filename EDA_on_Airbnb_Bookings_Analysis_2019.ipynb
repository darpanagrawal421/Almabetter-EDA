```python
new_col =df [['id', 'name', 'host_id', 'host_name', 'neighbourhood_group',
       'neighbourhood', 'room_type', 'price',
       'minimum_nights', 'number_of_reviews', 
       'calculated_host_listings_count',
       'availability_365']]

new_col.head()
```

**What can we learn about different hosts and areas?**

```python
hosts_areas = new_col.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].count().reset_index()
hosts_areas.sort_values(by='calculated_host_listings_count', ascending=False).head(5)
```

```python
area_reviews = new_col.groupby(['name', 'neighbourhood_group', 'price', 'minimum_nights'])['number_of_reviews'].count().reset_index()
price_reviews1 = area_reviews.sort_values(by = 'price', ascending = False).head(100)
price_reviews1
```

```python
area = price_reviews1['neighbourhood_group']
reviews = price_reviews1['price']

fig = plt.figure(figsize = (10, 5))

# creating the bar plot

plt.bar(area, reviews, color = 'orange', width = 0.4)

plt.xlabel('area')
plt.ylabel('price')
plt.title('Price vs Area')
plt.show()

```

**What can we learn from predictions? (ex: locations, prices, reviews, etc)**


```python
price_reviews = area_reviews.sort_values(by='number_of_reviews', ascending=False).head(100)
price_reviews
```

```python
reviews = price_reviews['number_of_reviews']
price = price_reviews['neighbourhood_group']

fig = plt.figure(figsize = (10, 5))
# creating the bar plot

plt.bar(price, reviews, color = 'maroon', width = 0.6)

plt.xlabel("neighbourhood_group")
plt.ylabel("reviews")
plt.title("neighbourhood vs number_of_reviews")
plt.show()
```

*Which hosts are the busiest and why?*

```python
busiest_hosts = new_col.groupby(['host_name','neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
busiest_hosts = busiest_hosts.sort_values(by='minimum_nights', ascending = False).head(10)
busiest_hosts
```

```python
name = busiest_hosts['host_name']
stayed = busiest_hosts['minimum_nights']

fig = plt.figure(figsize = (10,5))

# creating the bar plot

plt.bar(name, stayed, color = 'Green', width = 0.6)

plt.xlabel('Name of the Host')
plt.ylabel('Minimum number of nights stayed')
plt.title("Busiest Hosts")
plt.show()
```

Is there any noticeable difference of traffic among different areas and what could be the reason for it?

```python
traffic_areas = new_col.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_areas = traffic_areas.sort_values(by='minimum_nights', ascending = False).head(100)
traffic_areas
```

```python
room_type = traffic_areas['room_type']
stayed = traffic_areas['minimum_nights']

fig = plt.figure(figsize = (10,5))

# creating the bar plot

plt.bar(room_type, stayed, color = 'Red', width = 0.6)

plt.xlabel('Room Type')
plt.ylabel('Minimum number of nights stayed')
plt.title("Traffic Areas")
plt.show()
```
